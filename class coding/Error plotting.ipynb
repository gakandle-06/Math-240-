```python
# 1. Function and its derivative

x = 1  # Point where we differentiate
true_val = fp(x)

# 2. Step sizes
h = np.array([0.1 / (2**i) for i in range(5)])

# 3. Numerical differentiation
center, right, left = [], [], []

for step in h:
    c = (f(x + step) - f(x - step)) / (2 * step)   # Centered difference
    r = (f(x + step) - f(x)) / step                # Forward difference
    l = (f(x) - f(x - step)) / step                # Backward difference

    center.append(c)
    right.append(r)
    left.append(l)

# 4. Errors
center_error = np.abs(np.array(center) - true_val)
right_error = np.abs(np.array(right) - true_val)
left_error = np.abs(np.array(left) - true_val)

# 5. Plot on log-log scale
plt.figure(figsize=(8, 6))
plt.loglog(h, center_error, 'o-', label='Centered Difference Error')
plt.loglog(h, right_error, 's-', label='Forward Difference Error')
plt.loglog(h, left_error, '^-', label='Backward Difference Error')

plt.xlabel('Step size (h)')
plt.ylabel('Absolute Error')
plt.title('Differentiation Error vs Step Size')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function.1
def f(x):
    return (x**2) * np.log(x**3)  # natural log

# True derivative: f'(x) = 6xln(x) + 3x
def true_derivative(x):
    return 6*x*np.log(x) + 3*x

# Parameters
x_point = 1
h = 0.01

# Approximation functions
def left_diff(x, h):
    return (f(x) - f(x-h)) / h

def right_diff(x, h):
    return (f(x+h) - f(x)) / h

def center_diff(x, h):
    return (f(x+h) - f(x-h)) / (2*h)

# --- 1. Approximation of derivative at x=1 ---
center_val = center_diff(x_point, h)
right_val = right_diff(x_point, h)
left_val = left_diff(x_point, h)
true_val = true_derivative(x_point)

# --- 2. Errors ---
center_error = abs(center_val - true_val)
right_error = abs(right_val - true_val)
left_error = abs(left_val - true_val)

# Plot approximation at x=1
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.bar(["True", "Center", "Right", "Left"],
        [true_val, center_val, right_val, left_val],
        color=["black", "blue", "green", "red"])
plt.title(f"Derivative Approximation at x={x_point}, h={h}")
plt.ylabel("Value")

# Plot errors
plt.subplot(1, 3, 2)
plt.bar(["Center", "Right", "Left"],
        [center_error, right_error, left_error],
        color=["blue", "green", "red"])
plt.title("Approximation Errors")
plt.ylabel("Absolute Error")

# Plot derivative over [1,5]
x_vals = np.linspace(1, 5, 200)
true_vals = true_derivative(x_vals)
center_vals = [center_diff(x, h) for x in x_vals]

plt.subplot(1, 3, 3)
plt.plot(x_vals, true_vals, 'k-', label="True Derivative")
plt.plot(x_vals, center_vals, 'b--', label="Center Approximation")
plt.title(f"Derivative Approximation over [1,5], h={h}")
plt.xlabel("x")
plt.ylabel("Derivative")
plt.legend()

plt.tight_layout()
plt.show()
```
