```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2*np.cos(x)
```

```python
def p2(x):
    return x**2
```

```python
x = np.linspace(-(np.pi), np.pi,300)
```

```python
substed = np.absolute(f(x) - p2(x))
```

```python
plt.plot(x, substed)
plt.xlabel("x")
plt.ylabel("f(x) - p2(x)")
plt.title("Difference between f(x) = x²cos(x) and p2(x) = x²")
plt.grid(True)
plt.show()
```
