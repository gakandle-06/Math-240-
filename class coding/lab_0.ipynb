### Part 3

Define a function called `onlyPositive` that accepts a 1D array as input. 
This function will create a copy of the input array using the `copy()` method from Numpy, use Numpy's fancy indexing to set all negative values in array to 0, and finally return the modified copy.

````python
# x[[1,3]]
# x[[False, True, True, False]]
# x[[0,2,4]]
# x[0:5:2] #[::increment]
````

````python
# Define your function
def onlyPositive(x: np.array, copy: bool = False) -> np.array:
    # create a new array as a copy of the previous array
    #new_array = x.copy()
    '''
    Return a new array replacing the non-negative values of x with 0
    Args:
        x (numpy.array): input numpy array
        copy: if True, make a copy of x before modifying it

    Returns:
    u (numpy.array): A numpy array with the same shape as, where all negative values are replaced with 0.

    Example 
     >>>> x = np.array([3,-1,2,-5,0])
     >>>> onlypositive(x)
     >>>> [3,0,2,0,0]
    '''
    if copy:
        u = x.copy()
    else:
        u = x
    # set all negative values in new array to 0
    u[x<0] = 0
    # return the new array
    return x
onlyPositive(np.array([1,-1,1]))
````

## Exercise 4

Plot the function $f(x)=x^2 - 3$ on the interval $[-1, 10]$.

Add the legend using `legend()`, and add a title using `title()`.

````python
# Define your f function below
def f(x):
    return x**2 - 3
# Use the linspace function to define values of x below
x = np.linspace(-1,10,50)
y = f(x)

# Plot the function using pyplot below
plt.plot(x , y, label = "x vs y")

# Add a title
plt.title("f(x) = x^2 - 3")

# Add a x-label
plt.xlabel("x-axis")

# Add a y-label
plt.ylabel("y-axis")
plt.legend()
plt.show()
````
