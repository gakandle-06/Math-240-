```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2*np.log(x**3)

a=1
h=0.01
print((f(a+h)-f(a-h))/(2*h))
print((f(a)-f(a-h))/h)
print((f(a+h)-f(a))/h)
```

```python
# h = np.array([0.01/10**(-i) for i in range(5)])
h = np.array([0.01/2**(i) for i in range(5)])
exact=3
left = np.abs(exact-(f(a)-f(a-h))/h)
right = np.abs(exact-(f(a+h)-f(a))/h)
centered = np.abs(exact-(f(a+h)-f(a-h))/(2*h))

print('h=', h)
print('left=', left , 
      '\n\t slope =', np.log10(left[-1]/left[0])/np.log10(h[-1]/h[0]))
print('right=', right,
      '\n\t  slope =', np.log10(right[-1]/right[0])/np.log10(h[-1]/h[0]))
print('centered=', centered,
      '\n\t slope =', np.log10(centered[-1]/centered[0])/np.log10(h[-1]/h[0]))
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

ax[0].plot(h, left, label='left')
ax[0].plot(h, right, label='right')
ax[0].legend()

ax[1].plot(h, centered, 'g',label='centered')
ax[1].legend()
```

```python
plt.loglog(h, left, label='left')
plt.loglog(h, right, label='right')
plt.loglog(h, centered, label='centered')
plt.loglog(h, h, 'k--', label='slope 1')
plt.loglog(h, h**2, 'k:', label='slope 2')
plt.legend()
plt.show()
```
