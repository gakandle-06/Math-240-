## Exercise 6
Let us compute $\sqrt[m]{a}$, with $a>0$ and $m$ a positive integer. Find a way to
apply the Newton's method to this problem to find $\sqrt[m]{5}$ for
$m=3,4,5,6,7,8,$ say to six significant digits.

$x = a^\frac{1}{m}$

$x^m = a \longrightarrow x^m - a = 0$

$a = 5 \Rightarrow x^m - 5 = 0$

````python
m_vals = [3, 4, 5, 6, 7, 8]
for m in m_vals:
    print(f"\nCalculating for m = {m}")
    f = lambda x: x**m - 5
    df = lambda x: m*x**(m-1)
    tol = 1e-7
    approx_xm, _, _= Newton(f, df, 1, tol=tol)
    print(f"Root: {approx_xm:.6f}")
    print(f"Verification: {(5**(1/m)):.6f}")
````

## Exercise 8
Using the Newton's method Find the minimum value of
$$(1-x)^2+e^{x^2}-2\log( (3-x)^2)$$

Finding the minimum value by finding the roots of f prime.

**So in Newton method `f(x) = f'(x)` and `f'(x) = f''(x)`**

````python
z = sp.symbols('z')
````

````python
f_sym = (1-z)**2 + sp.exp(z**2) - 2*sp.log((3-z)**2)
df_sym = sp.diff(f_sym, z, 1)
dff_sym = sp.diff(df_sym, z, 1)

f_num = sp.lambdify(z, f_sym, 'numpy')
df_num = sp.lambdify(z, df_sym, 'numpy')
dff_num = sp.lambdify(z, dff_sym, 'numpy')
````

````python
minimum_z, _, _ = Newton(df_num, dff_num, 0.5)
minimum_value = f_num(minimum_z)
print(f"The minimum occurs at x ≈ {minimum_z:.8f}")
print(f"The minimum value of f(x) ≈ {minimum_value:.8f}")
````

We can convert the equation $x^2-6=0$ into a fixed-point problem:
$$x=x+c(x^2-6)$$
Find a $c$ such that the fixed point methods converges to the solution.

**The method of fix**: You want to find g(x)=x, you perform
$$x_{n+1}=g(x_n).$$
You can use this method to find roots as well. Assume you want f(x)=0, you define
g(x)=x+f(x) then the fixed point of g is the root of f
$$g(x)=x \Longleftrightarrow x+f(x)=x \Longleftrightarrow f(x)=0$$

------
The method converges if the absolute value of the derivative, $|g'(x)|$, is less than 1 at the root.

so this means that:
- if $|g'(x)|$ < 1, the process **converges**, and vice-versa when $|g'(x)|$ > 1 (**diverges**)

### Convergence Condition at root

$x = \sqrt6 \approx 2.449$

$|g'(x)| < 1 \Longrightarrow |1 + c(2x)|$

#### Solving inequality, Substituting x, and so on
$-1< 1 + c (2 \sqrt 6)< 1$

$-2< 2c \sqrt 6 < 0$

$\frac{-2}{2\sqrt 6}< |c| < \frac{0}{2\sqrt 6}$
$\frac{-2}{2\sqrt 6}< |c| < 0 \Longleftrightarrow \frac{-1}{\sqrt 6}< |c| < 0$



So, for convergence to sqrt(6), we need c in $(\frac{-1}{\sqrt 6}, 0)$.
For convergence to -sqrt(6), we need c in $(0, \frac{-1}{\sqrt 6})$.


````python
c_pos = -0.1 # Should converge to sqrt(6)
c_neg = 0.1  # This Should converge to -sqrt(6)
x0 = 1.0

g = lambda x, c: x + c*(x**2 - 6)
g_prime = lambda x, c: 1 + c*(2*x)
tol = 1e-8

true_result = np.sqrt(6)
````

````python
print(f"Testing convergence for c = {c_pos}")
print(f"Expecting sqrt(6) ≈ {np.sqrt(6):.6f}")
xn = x0 
# Newton(g,  g_prime, x0, tol=tol)
for i in range(25):
    xn_1 = g(xn, c_pos)
    if abs(xn_1 - xn) < tol: break
    xn = xn_1
print(f"\tConverged to {xn:.8f} in {i+1} iterations\n")

print(f"Testing convergence for c = {c_neg}")
print(f"Expecting sqrt(6) ≈ -{np.sqrt(6):.6f}")
xn = x0
for i in range(25):
    xn_1 = g(xn, c_neg)
    if abs(xn_1 - xn) < tol: break
    xn = xn_1
print(f"\tConverged to {xn:.8f} in {i+1} iterations")
````
