```python
def gradient_descent(f_prime, x0, a, max_iter, error):
    x = x0
    for i in range(max_iter): 
        xnew = x - a*f_prime(x)
        if np.abs(xnew - x) < error:
            print(f"Converged at iteration {i+1}: x = {xnew}")
            return xnew, i + 1
        x = xnew
        if i % 10 == 0:
            print(f"Iter {i}: x = {x}")
    print(f"\nReached max iterations ({max_iter}): x = {x}") 
    return xnew, i+1
```
